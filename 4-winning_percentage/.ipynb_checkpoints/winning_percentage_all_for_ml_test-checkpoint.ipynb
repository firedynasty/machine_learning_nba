````python
df_into_season = df_2.copy()
df_into_season['DATE'] = pd.to_datetime(df_into_season['DATE'])
````

````python
empty_home_winning_percentage = []
empty_away_winning_percentage = []
home_team_wins = 0
home_team_loss = 0
away_team_wins = 0
away_team_loss = 0
for ind in df_into_season.index:
    if ind > 77:
        home_team_wins = 0
        home_team_loss = 0
        away_team_wins = 0
        away_team_loss = 0
        date = df_into_season['DATE'][ind]
        home_team_abbr = df_into_season['HOME_ABBR'][ind]
        visitor_team_abbr = df_into_season['VISITOR_ABBR'][ind]
        home_team_wins = df_into_season['HOME_TEAM_WINS'][ind]
        mask = (df_into_season['DATE'] < date)
        new_ = df_into_season.loc[mask]
        home_df2 = new_.loc[new_['HOME_ABBR'] == home_team_abbr]
        home_as_visitor_df2 = new_.loc[new_['VISITOR_ABBR'] == home_team_abbr]
        visitor_df2= new_.loc[new_['VISITOR_ABBR'] == visitor_team_abbr]
        visitor_as_home_df2 = new_.loc[new_['HOME_ABBR'] == visitor_team_abbr]
        for ind in home_df2.index:
            home_team_wins_home_df2 = home_df2['HOME_TEAM_WINS'][ind]
            if home_team_wins_home_df2 == 1:
                home_team_wins = home_team_wins + 1
            else:
                home_team_loss = home_team_loss + 1
        for ind in home_as_visitor_df2.index:
            home_team_wins_visitor_df2 = home_as_visitor_df2['HOME_TEAM_WINS'][ind]
            if home_team_wins_visitor_df2 == 0:
                home_team_wins = home_team_wins + 1
            else:
                home_team_loss = home_team_loss + 1
    
        for ind in visitor_df2.index:
            home_team_wins_visitor_df2 = visitor_df2['HOME_TEAM_WINS'][ind]
#     print(visitor_df2['HOME_TEAM_WINS'][ind])
            if home_team_wins_visitor_df2 == 0:
                away_team_wins = away_team_wins + 1
            else:
                away_team_loss = away_team_loss + 1
        for ind in visitor_as_home_df2.index:
            home_team_wins_visitor_as_home_df2 = visitor_as_home_df2['HOME_TEAM_WINS'][ind]
            if home_team_wins_visitor_as_home_df2 == 1:
                away_team_wins = away_team_wins + 1
            else:
                away_team_loss = away_team_loss + 1            
        total_home_team_games = home_team_wins + home_team_loss
        total_away_team_games = away_team_wins + away_team_loss
        home_winning_percentage = home_team_wins / total_home_team_games
        away_winning_percentage = away_team_wins / total_away_team_games
        empty_home_winning_percentage.append(home_winning_percentage)
        empty_away_winning_percentage.append(away_winning_percentage)
````

````python
import numpy as np
length_beg_arr = len(df_into_season) - len(empty_away_winning_percentage)
arr = np.arange(78)
arr[:] = 0.0 
arr = arr.tolist()
empty_away_winning_percentage = arr + empty_away_winning_percentage
empty_home_winning_percentage = arr + empty_home_winning_percentage
````

````python
df_into_season['home_winning_percentage'] = empty_home_winning_percentage
df_into_season['visitor_winning_percentage'] = empty_away_winning_percentage
````

````python
# create dataframe with respective columns
# turn this into a dictionary to add to the rows that have 0.0 for winning_percentage
team_abbr = df_into_season['HOME_ABBR'].unique()
team_abbr = team_abbr.tolist()
team_abbr_df = pd.DataFrame(team_abbr)
team_abbr_df.rename(columns={0:'team_abbr'}, inplace=True)
team_abbr_df['winning_percentage'] = 0.0
````

````python
team_winning_dict = dict(zip(team_abbr_df['team_abbr'], team_abbr_df['winning_percentage']))
````

````python
for ind in df_into_season.index:
    abbr = df_into_season['HOME_ABBR'][ind]
    winning_percentage = df_into_season['home_winning_percentage'][ind]
    team_winning_dict[abbr] = winning_percentage
````
