````python
margin = df['MARGIN']
assists = df['AST']
plt.scatter(margin, assists)
plt.xlabel('margin')
plt.ylabel('assists')
plt.show()
````

````python
import scipy.stats as st
cor = round(st.pearsonr(margin, assists)[0], 2)
print(f'Correleation between margin and assists is {cor}')
model = st.linregress(margin, assists)
y_values = margin * model[0] + model[1]
plt.scatter(margin, assists)
plt.plot(margin, y_values, color="red")
plt.xlabel('margin')
plt.ylabel('assists')
plt.show()
````

````python
#Graphing 
point = (new_stats["FGA"])
points = point.dropna()
minute = new_stats["PTS"]
minutes = minute.dropna()
#plt.figure(figsize=(20, 30))
plt.scatter(points, minutes)
plt.xlabel("Minutes")
plt.ylabel("Game Score")
plt.show()
````

````python
#Linear Regression
import scipy.stats as st
cor=round(st.pearsonr(points,minutes)[0],2)
print(f"The correlation between points and minutes is {cor}")
model = st.linregress(points,minutes)

y_values = points*model[0]+model[1]
plt.scatter(points,minutes)
plt.plot(points,y_values,color="red")
plt.xlabel("points")
plt.ylabel('minutes')
plt.show()
````
