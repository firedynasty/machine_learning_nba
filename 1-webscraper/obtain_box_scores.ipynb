```python
box_score_columns = box_score_columns = [ 'DATE', 'HOME_ABBR', 'VISITOR_ABBR', 'PTS_home', 'PTS_away', 'MARGIN', 'HOME_TEAM_WINS', 'FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away', 'AST_away', 'REB_away']
box_score_df = pd.DataFrame(columns=box_score_columns)


from basketball_reference_scraper.box_scores import get_box_scores

import time
for ind in schedule_df.index:
    date_1 = schedule_df.loc[ind]['DATE']
    home_1 = schedule_df.loc[ind]['HOME_ABBR']
    visitor_1 = schedule_df.loc[ind]['VISITOR_ABBR']
    dict_ = get_box_scores(date_1, visitor_1, home_1, period='GAME', stat_type='BASIC')
    home_df = dict_[home_1]
    visitor_df = dict_[visitor_1]
    #id_ = schedule_df.loc[ind]['ID']
    #unsure why the ID is required
    home_pts = home_df.loc[home_df['PLAYER'] == 'Team Totals']['PTS'].iloc[0]
    visitor_pts = visitor_df.loc[visitor_df['PLAYER'] == 'Team Totals']['PTS'].iloc[0]
    margin_ = int(home_pts) - int(visitor_pts)
    if margin_ > 0:
        home_team_wins = 1
    else:
        home_team_wins = 0
    home_fg_pct = home_df.loc[home_df['PLAYER'] == 'Team Totals']['FG%'].iloc[0]
    home_ft_pct = home_df.loc[home_df['PLAYER'] == 'Team Totals']['FT%'].iloc[0]
    home_fg3_pct = home_df.loc[home_df['PLAYER'] == 'Team Totals']['3P%'].iloc[0]
    home_ast = home_df.loc[home_df['PLAYER'] == 'Team Totals']['AST'].iloc[0]
    home_reb = home_df.loc[home_df['PLAYER'] == 'Team Totals']['TRB'].iloc[0]
    visitor_fg_pct = visitor_df.loc[visitor_df['PLAYER'] == 'Team Totals']['FG%'].iloc[0]
    visitor_ft_pct = visitor_df.loc[visitor_df['PLAYER'] == 'Team Totals']['FT%'].iloc[0]
    visitor_fg3_pct = visitor_df.loc[visitor_df['PLAYER'] == 'Team Totals']['3P%'].iloc[0]
    visitor_ast = visitor_df.loc[visitor_df['PLAYER'] == 'Team Totals']['AST'].iloc[0]
    visitor_reb = visitor_df.loc[visitor_df['PLAYER'] == 'Team Totals']['TRB'].iloc[0]
    empty_ = []
    #empty_.append(id_)
#ID unsure why that is required
    empty_.append(date_1)
#DATE
    empty_.append(home_1)
#home abbr
    empty_.append(visitor_1)
#visito abbr
    empty_.append(home_pts)
#home pints
    empty_.append(visitor_pts)
#away points
    empty_.append(margin_)
#margin
    empty_.append(home_team_wins)
#HOME_TEAM_WINS
    empty_.append(home_fg_pct)
#FG_PCT
    empty_.append(home_ft_pct)
#FT_PCT
    empty_.append(home_fg3_pct)
#FG3_PCT
    empty_.append(home_ast)
#AST
    empty_.append(home_reb)
#REBOUNDS
    empty_.append(visitor_fg_pct)
#FG_PCT
    empty_.append(visitor_ft_pct)
#FT_PCT
    empty_.append(visitor_fg3_pct)
#FG3
    empty_.append(visitor_ast)
#AST
    empty_.append(visitor_reb)
#REBOUNDS
    length_of_df = len(box_score_df)
    box_score_df.loc[length_of_df] = empty_
    print(empty_)

```

```python
home_winning_percentage_list = []
visitor_winning_percentage_list = []
for ind in box_score_df.index:
    home_abbr = box_score_df.loc[ind]['HOME_ABBR']
    visitor_abbr = box_score_df.loc[ind]['VISITOR_ABBR']
    home_team_wins = box_score_df.loc[ind]['HOME_TEAM_WINS']
    if home_team_wins > 0:
        wins_[home_abbr] += 1
        lost_[visitor_abbr] += 1
    else:
        lost_[home_abbr] += 1
        wins_[visitor_abbr] += 1
    total_games_home_abbr = wins_[home_abbr] + lost_[home_abbr]
    total_games_visitor_abbr = wins_[visitor_abbr] + lost_[visitor_abbr]
    home_winning_percentage = wins_[home_abbr] / total_games_home_abbr
    visitor_winning_percentage = wins_[visitor_abbr] / total_games_visitor_abbr
    home_winning_percentage_list.append(home_winning_percentage)
    visitor_winning_percentage_list.append(visitor_winning_percentage)
box_score_df['home_winning_percentage'] = home_winning_percentage_list
box_score_df['visitor_winning_percentage'] = visitor_winning_percentage_list
```
