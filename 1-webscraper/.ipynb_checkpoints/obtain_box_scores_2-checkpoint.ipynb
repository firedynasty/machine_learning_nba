```python
#based on the group_by_date_for_ml you want to see what is the differences that come up
# based on winning percentage differential

diff_df_columns = [ 'DATE', 'HOME_ABBR', 'VISITOR_ABBR', 'PTS_home', 'PTS_away', 'MARGIN', 'FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away']
diff_df = pd.DataFrame(columns=diff_df_columns)

for ind in group_by_date_for_ml.index:
    empty_ = []
    home_abbr = group_by_date_for_ml.loc[ind]['HOME_ABBR']
    visitor_abbr = group_by_date_for_ml.loc[ind]['VISITOR_ABBR']
    diff = group_by_date_for_ml.loc[ind]['diff_%']
    home_df = calculated_df[calculated_df['HOME_ABBR'] == home_abbr]
    home_df['diff_%'] = home_df['home_winning_percentage'] - home_df['visitor_winning_percentage']
    home_df = home_df.round({"diff_%":1})
    mean_grouped_home_df = home_df.groupby("diff_%").mean().reset_index()
    list_ = mean_grouped_home_df['diff_%'].tolist()
    arr = np.asarray(list_)
    input_val = (np.abs(arr - 0)).argmin()
    return_ = arr[input_val]
    if (diff in list_):
        mean_grouped_home_df = mean_grouped_home_df.loc[mean_grouped_home_df['diff_%'] == diff]
    else:
        mean_grouped_home_df = mean_grouped_home_df.loc[mean_grouped_home_df['diff_%'] == return_]
        print(f'{home_abbr} does not have the average correct')
        print(f'Ind:{ind}')
    mean_grouped_home_df = mean_grouped_home_df.loc[mean_grouped_home_df['diff_%'] == diff]
    visitor_df = calculated_df[calculated_df['VISITOR_ABBR'] == visitor_abbr]
    visitor_df['diff_%'] = visitor_df['home_winning_percentage'] - visitor_df['visitor_winning_percentage']
    visitor_df = visitor_df.round({"diff_%":1})
    mean_grouped_visitor_df = visitor_df.groupby("diff_%").mean().reset_index()
    list_ = mean_grouped_visitor_df['diff_%'].tolist()
    arr = np.asarray(list_)
    input_val = (np.abs(arr - 0)).argmin()
    return_ = arr[input_val]
    if (diff in list_):
        mean_grouped_visitor_df = mean_grouped_visitor_df.loc[mean_grouped_visitor_df['diff_%'] == diff]
    else:
        mean_grouped_visitor_df = mean_grouped_home_df.loc[mean_grouped_home_df['diff_%'] == return_]
        print(f'{visitor_abbr} does not have an average group')
    mean_grouped_visitor_df = mean_grouped_visitor_df.loc[mean_grouped_visitor_df['diff_%'] == diff]
    grouped_df = pd.concat((mean_grouped_home_df, mean_grouped_visitor_df))
    empty_.append(group_by_date_for_ml.loc[ind]['DATE'])
    empty_.append(home_abbr)
    empty_.append(visitor_abbr)
    empty_.append(grouped_df.mean()['PTS_home'])
    empty_.append(grouped_df.mean()['PTS_away'])
    empty_.append(grouped_df.mean()['MARGIN'])
    empty_.append(grouped_df.mean()['FG_PCT_home'])
    empty_.append(grouped_df.mean()['FT_PCT_home'])
    empty_.append(grouped_df.mean()['FG3_PCT_home'])
    empty_.append(grouped_df.mean()['FG_PCT_away'])
    empty_.append(grouped_df.mean()['FT_PCT_away'])
    empty_.append(grouped_df.mean()['FG3_PCT_away'])
    length_of_df = len(diff_df)
    diff_df.loc[length_of_df] = empty_
```
