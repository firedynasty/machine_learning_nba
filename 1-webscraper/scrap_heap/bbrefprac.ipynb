```python
#Clean
#clean_stats = stats_df['PTS'] = pd.to_numeric(df['MP'], errors='coerce')
new_stats = pd.read_csv(r'C:\Users\Cody\UCBE\final\stats.csv')
new_stats.head()
```

```python
#Graphing 
point = (new_stats["FGA"])
points = point.dropna()
minute = new_stats["PTS"]
minutes = minute.dropna()
#plt.figure(figsize=(20, 30))
plt.scatter(points, minutes)
plt.xlabel("Minutes")
plt.ylabel("Game Score")
plt.show()
```

```python
#Linear Regression
import scipy.stats as st
cor=round(st.pearsonr(points,minutes)[0],2)
print(f"The correlation between points and minutes is {cor}")
model = st.linregress(points,minutes)

y_values = points*model[0]+model[1]
plt.scatter(points,minutes)
plt.plot(points,y_values,color="red")
plt.xlabel("points")
plt.ylabel('minutes')
plt.show()
```
