```python
CREATE TABLE predicted_values (
    "ID" INT PRIMARY KEY, 
    "DATE" DATE,
    "HOME_ABBR" VARCHAR,
    "VISITOR_ABBR" VARCHAR,
    "MARGIN" INT,
    "PTS_home" INT,
    "PTS_away" INT,
    "FG_PCT_home" DOUBLE PRECISION,
    "FG_PCT_away" DOUBLE PRECISION,
    "PREDICTED" DOUBLE PRECISION,
    "CONVERTED_PREDICTED" DOUBLE PRECISION,
    "WINNING_PCT" DOUBLE PRECISION
);
```
