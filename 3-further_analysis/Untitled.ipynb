```python
import pandas as pd
box_score = pd.read_csv('03_03box_score_combined.csv')
```

```python
calculated_df = pd.read_csv('03_03_calculated_df_full.csv')
```

```python
box_score['diff_%'] = box_score['home_winning_percentage'] - box_score['visitor_winning_percentage']
calculated_df['diff_%'] = calculated_df['home_winning_percentage'] - calculated_df['visitor_winning_percentage']
```

```python
box_score_2 = box_score[['DATE', 'HOME_ABBR', 'VISITOR_ABBR', 'PTS_home', 'PTS_away', 'MARGIN',
       'HOME_TEAM_WINS', 'home_winning_percentage', 'visitor_winning_percentage', 'diff_%']]
```

```python
calculated_df_2 = calculated_df[['DATE', 'HOME_ABBR', 'VISITOR_ABBR', 'PTS_home', 'PTS_away', 'MARGIN', 'home_winning_percentage', 'visitor_winning_percentage', 'diff_%']]
```

```python
home_df = calculated_df_2[calculated_df_2['HOME_ABBR'] == home_user_input]
home_df = home_df.round({"diff_%":1})
mean_grouped_home_df = home_df.groupby("diff_%").mean().reset_index()
mean_grouped_home_df
```

```python
visitor_df = calculated_df_2[calculated_df_2['VISITOR_ABBR'] == visitor_user_input]
visitor_df = visitor_df.round({"diff_%":1})
mean_grouped_visitor_df = visitor_df.groupby("diff_%").mean().reset_index()
mean_grouped_visitor_df
```
