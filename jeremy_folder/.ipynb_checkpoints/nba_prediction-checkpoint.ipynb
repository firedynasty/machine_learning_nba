```python
# get a team's past 10 games from a given date:

#let's look at Golden State's 73-9 record from the 2015-2016 season
season_2016_df = get_schedule(2016, playoffs=False)

season_2016_df
```

```python
#their largest loss was against the Blazers, on 2016-02-19, so we'll grab all the data of warrior games before this date
season_2016_pre_big_loss = season_2016_df.loc[(season_2016_df["DATE"] < "2016-02-19")]
#a line that didnt work: (team_to_abbreviation[str(season_2016_df["VISITOR"]).upper()] == "GSW" or team_to_abbreviation[str(season_2016_df["HOME"]).upper()] == "GSW")

# warriors_2016_pre_big_loss =  season_2016_pre_big_loss.loc[(team_to_abbreviation[season_2016_df["VISITOR"]] == "Golden State Warriors" | team_to_abbreviation[season_2016_df["HOME"]] == "Golden State Warriors")]


```

```python
import pandas as pd
# warriors_2016_pre_big_loss = pd.DataFrame()

# for rows in season_2016_pre_big_loss:
#     if ((rows["HOME"] == "Golden State Warriors") or (rows ["VISITOR"] == "Golden State Warriors")):
#         warriors_2016_pre_big_loss.append(rows)
        
warriors_2016_home = season_2016_pre_big_loss.loc[(season_2016_df["HOME"] == "Golden State Warriors")]
warriors_2016_visitor = season_2016_pre_big_loss.loc[(season_2016_df["VISITOR"] == "Golden State Warriors")]

season = [warriors_2016_home, warriors_2016_visitor]

warriors_season = pd.concat(season)

warriors_season=  warriors_season.sort_values(by="DATE", ascending = False)

# warriors_season
last_10_games = pd.DataFrame(warriors_season[0:11])
last_10_games
```

```python
last_10_games_stats = []
game_date = last_10_games.iloc[0]["DATE"]
date_no_time = str(game_date).split()[0]
date_no_time

visitor_team = team_to_abbreviation[last_10_games.iloc[0]["VISITOR"].upper()]
visitor_team

home_team = team_to_abbreviation[last_10_games.iloc[0]["HOME"].upper()]
home_team

team_box_score = get_box_scores(date_no_time, visitor_team, home_team, period='GAME', stat_type='BASIC')
team_box_score = pd.DataFrame(team_box_score["GSW"])

team_box_score.tail(1)


# for index, row in last_10_games.iterrows():
#     print("hi")
```
